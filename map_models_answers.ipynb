## Mapeando os gabaritos retornados dos modelos UI

```python
import os
import pandas as pd
CSV_PATH = "../TCC/csvs"
EXPERIMENTS_PATH = '../TCC/experiments'
```

##### Prenecha o número do experimento aqui

```python
EXPERIMENT_NUMBER = 2
```

```python
def format_Llama_answers_txt(answers_txt):
    lines = answers_txt.split('\n')
    result = ''

    for line in lines:
        line = line.replace('"', '')
        line = line.replace(':', '')
        if not line.strip().isalnum() and line.strip() != '' :
            index = line.find('é')
            result += line[index+2] + '\n'
            continue

        result += line + '\n'
    return result
```

```python
def format_GPT_answers_txt(answers_txt):
    
    lines = answers_txt.split('\n')
    result = ''
    print(lines)
    for line in range(0,len(lines),2):
        result += lines[line] + '\n' + lines[line+1][0] + '\n'
        
    return result
```

```python
def format_GEMINI_answers_txt(answers_txt):
    answers = answers_txt.split("\n")
    result = ''
    
    for line in answers:
        
        if not line.strip().isalnum() and line.strip() != '' :
            
            if '**' in line:
                index = line.find('**')
                
                result += line[index+2] + '\n'
            else:
                # print(line[0])
                result += line[0] + '\n'
        
        else:
            result += line + '\n'
                
    return result
```

```python
def format_answers_txt(filename, answers_txt):
    if( 'GEMINI' in filename ):
        return format_GEMINI_answers_txt(answers_txt)
    elif( 'GPT' in filename):
        return format_GPT_answers_txt(answers_txt)
    elif( 'Llama' in filename):
        return format_Llama_answers_txt(answers_txt)
```

```python
def map_answers(answers_key_txt):
    answers_key_lines = answers_key_txt.split('\n')
            
    answers_map = {}
    for i in range(0, len(answers_key_lines)-1, 2):
        key = answers_key_lines[i]
        value = answers_key_lines[i + 1]
        
        answers_map[key] = value if len(value) > 0 else 'No answer'  
             
    return answers_map
```

```python
def aggregate_models_answers(answers_map, year, experiment_name):
    df = pd.read_csv(f'{EXPERIMENTS_PATH}/{EXPERIMENT_NUMBER}/{year}/enem_{year}.csv')
    df[experiment_name] = df['id'].map(answers_map)
    # print(len(answers_map))
    print(answers_map)
    display(df.head())
    return df
```

```python
years = os.listdir(f'{EXPERIMENTS_PATH}/{EXPERIMENT_NUMBER}')
years = [2013]
for year in years:
    
    experiments = os.listdir(f'{EXPERIMENTS_PATH}/{EXPERIMENT_NUMBER}/{year}')
    
    df = pd.read_csv(f'{CSV_PATH}/enem_{year}_questions.csv')
    df.to_csv(f'{EXPERIMENTS_PATH}/{EXPERIMENT_NUMBER}/{year}/enem_{year}.csv', index=False)
    
    for experiment in experiments:
        
        if '.txt' not in experiment:
            continue
        
        print(experiment)
        
        experiment_name = experiment.split('.txt')[0]
        
        with open(f'{EXPERIMENTS_PATH}/{EXPERIMENT_NUMBER}/{year}/{experiment}', 'r', encoding='utf-8') as exam_str:
            
            content = exam_str.read()
            content = format_answers_txt(experiment_name, content)
            answers_map = map_answers(content)
            df_with_answers = aggregate_models_answers(answers_map, year,experiment_name)
        df_with_answers.to_csv(f'{EXPERIMENTS_PATH}/{EXPERIMENT_NUMBER}/{year}/enem_{year}.csv', index=False)

```
